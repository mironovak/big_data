Также выведите названия кафедр, которые совместно преподают один и тот же курс.




```python
# Ищем курсы с >1 кафедрой

#TRING_AGG(DISTINCT d.name, ', ') — функция, которая объединяет строки в одну
#WHERE u.courseid IN ( ... ) - Подзапрос, который выбирает только те курсы, у которых количество уникальных кафедр больше 1.
#
cursor.execute("""
SELECT u.courseid, STRING_AGG(DISTINCT d.name, ', ') as departments
FROM user_logs u
JOIN departments d ON u.depart = d.id
WHERE u.courseid IN (
    SELECT courseid 
    FROM user_logs 
    GROUP BY courseid 
    HAVING COUNT(DISTINCT depart) > 1
)
GROUP BY u.courseid;
""")

results = cursor.fetchall()
print(f"Количество таких курсов: {len(results)}")
#Проверяет, есть ли хотя бы одна строка в results. Если список не пуст, выполняет код внутри.
if results:
    print("ID курса - Кафедры")
    for row in results:
        #Выводит первый элемент (ID курса) и второй (строку с кафедрами), разделённые " - ".
        print(f"{row[0]} - {row[1]}")
```

## Задание 6:

Выведите студента, который больше всех работает на портале (у него максимальное количество логов за вест период обучения).

```python
#Поскольку s_all — накопительное количество событий за весь период (оно увеличивается с каждой неделей), 
#MAX(s_all) фактически равно суммарному количеству логов студента за семестр.
cursor.execute(''' 
    SELECT userid
    FROM user_logs
    GROUP BY userid
    ORDER BY MAX(s_all) DESC
    LIMIT 1;

''')
row = cursor.fetchone()
print(f"Студент с максимальной активностью: id {row[0]}")
```

Отдельно выведите название кафедры, у которой больше всего двоечников.

```python
# Больше всего отличников

#COUNT(DISTINCT u.userid) — подсчитывает количество уникальных студентов. DISTINCT необходим, потому что один студент может иметь несколько записей логов с оценкой 5 (за разные недели). 
#Нам нужно число студентов
#ORDER BY student_count DESC - Сортирует группы по убыванию количества студентов 
cursor.execute("""
SELECT d.name, COUNT(DISTINCT u.userid) as student_count
FROM user_logs u
JOIN departments d ON u.depart = d.id
WHERE u.namer_level = '5'
GROUP BY d.name
ORDER BY student_count DESC
LIMIT 1;
""")
print(f"Больше всего отличников на кафедре: {cursor.fetchone()[0]}")
# Больше всего двоечников
cursor.execute("""
SELECT d.name, COUNT(DISTINCT u.userid) as student_count
FROM user_logs u
JOIN departments d ON u.depart = d.id
WHERE u.namer_level = '2'
GROUP BY d.name
ORDER BY student_count DESC
LIMIT 1;
""")
print(f"Больше всего двоечников на кафедре: {cursor.fetchone()[0]}")
```

## Задание 9:
Провести анализ пиковой активности студентов перед экзаменом (с использованием (Common Table Expression — CTE), оператор with).

Вывести, на какой неделе семестра студенты проявляли наибольшую активность в курсе в целом, и как эта активность распределяется между студентами-бюджетниками и контрактниками.

Пример вывода :

| name_osno | week_number	| avg_s_all	| avg_s_course_viewed |	avg_s_q_attempt_viewed |
|-----|------|------|------|------|
| бюджет |	14	| 125.45 |	45.67 |	32.12 |
|контракт |	14	| 98.76 |	38.90 |	25.43 |

```python
import pandas as pd
from IPython.display import display
#WITH Activity AS — создаёт временное именованное представление (CTE) с именем Activity. Оно существует только в рамках этого запроса
#CASE name_osno WHEN '1' THEN 'бюджет' WHEN '2... - Преобразует код формы обучения (предположительно '1' для бюджета, '2' для контракта) в читаемый текст.
#AVG(s_all) as avg_s_all — среднее всех событий (по группе: форма обучения + неделя).
#AVG(s_course_viewed) as avg_course_viewed — среднее просмотров курса.
#AVG(s_q_attempt_viewed) as avg_q_attempt — среднее просмотров тестов.

#SELECT name_osno_text, num_week, avg_s_all, avg_course_viewed, avg_q_attempt FROM Activity — выбирает все столбцы из CTE.
#ORDER BY avg_s_all DESC — сортирует по среднему количеству всех событий по убыванию.
#GROUP BY name_osno, num_week — группировка по форме обучения и неделе. Внутри каждой такой группы вычисляются средние значения.
cursor.execute(''' 
    WITH Activity AS (
        SELECT 
               CASE name_osno
                   WHEN '1' THEN 'бюджет' 
                   WHEN '2' THEN 'контракт' 
                   ELSE name_osno
               END as name_osno_text, 
               num_week, 
               AVG(s_all) as avg_s_all, 
               AVG(s_course_viewed) as avg_course_viewed, 
               AVG(s_q_attempt_viewed) as avg_q_attempt
        FROM user_logs
        GROUP BY name_osno, num_week
    )
    SELECT name_osno_text, num_week, avg_s_all, avg_course_viewed, avg_q_attempt
    FROM Activity
    ORDER BY avg_s_all DESC
    LIMIT 10;
''')
rows = cursor.fetchall()
columns = ["name_osno", "week_number", "avg_s_all", "avg_s_course_viewed", "avg_s_q_attempt_viewed"]
df = pd.DataFrame(rows, columns=columns)
#df.sort_values(by="week_number", ascending=True) — сортирует DataFrame по столбцу week_number в порядке возрастания (1,2,3...).
#.reset_index(drop=True) — сбрасывает индекс после сортировки, чтобы нумерация строк шла с 0 по порядку, без сохранения старого индекса.
df = df.sort_values(by="week_number", ascending=True).reset_index(drop=True)
display(df)
```
